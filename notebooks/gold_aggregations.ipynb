# Gold (Business Layer)

**Create curated tables:**

- daily_sales_metrics
- labour_cost_vs_revenue
- inventory_waste_analysis
- location_performance


```python
from pyspark.sql.functions import sum

revenue = spark.table("silver_sales") \
    .withColumn("revenue", col("quantity") * col("price")) \
    .groupBy("date", "location_id") \
    .agg(sum("revenue").alias("total_revenue"))

labour = spark.table("silver_labour") \
    .groupBy("date", "location_id") \
    .agg(sum("labour_cost").alias("total_labour_cost"))

final_df = revenue.join(labour, ["date", "location_id"]) \
    .withColumn("profit", col("total_revenue") - col("total_labour_cost"))

final_df.write.mode("overwrite").saveAsTable("gold_daily_metrics")
```
